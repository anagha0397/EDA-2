# One Hot Encoding

```python
data2 = pd.read_csv("iris.csv",index_col = 0)
```

# Isolation Forest

```python
data_iso =pd.read_csv("iris.csv",index_col = 0)
```

```python
data_dummy = pd.get_dummies(data_iso)
```

```python
model_iso = IsolationForest(random_state = 10, contamination = 0.01)  # Contamination is a amount of percentage which we want to consider in finding the outliers
model_iso.fit(data_dummy)
```

```python
preds_iso = model_iso.predict(data_dummy)
```

```python
preds_iso
```

```python
# Creating a new outlier in the data 
```

```python
data_dummy.loc[150] = [20, 30, 40, 50, 0,1,0]
```

```python
data_dummy
```

```python
model_iso = IsolationForest(random_state = 10, contamination = 0.01)  
model_iso.fit(data_dummy)
```

```python
preds_iso = model_iso.predict(data_dummy)
```

```python
preds_iso
```

```python
data_dummy['scores'] = model_iso.decision_function(data_dummy)
data_dummy['anomaly'] = model_iso.predict(data_dummy.iloc[:,0:7])
```

```python
data_dummy
```

```python
data_dummy[data_dummy['anomaly']==-1]
```

From this we can infer that there are outliers present in the 107th and 150th records
